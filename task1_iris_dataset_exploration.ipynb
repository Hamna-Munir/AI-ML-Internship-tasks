```python
# Import required libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Set plot style
sns.set(style='whitegrid')
%matplotlib inline
```

```python
# Load Iris dataset from seaborn
iris = sns.load_dataset("iris")

# Show basic info
print("Shape of dataset:", iris.shape)
print("\nColumn names:", iris.columns.tolist())

# First 5 rows
iris.head()
```

```python
# Info about data types and nulls
iris.info()

# Descriptive statistics
iris.describe()
```

```python
sns.pairplot(iris, hue='species')
plt.suptitle("Pairwise Scatter Plots by Species", y=1.02)
plt.show()
```

```python
iris.hist(figsize=(10, 8), color='skyblue', edgecolor='black')
plt.suptitle("Histograms of Iris Features")
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(data=iris)
plt.title("Boxplot of Iris Features")
plt.show()
```
